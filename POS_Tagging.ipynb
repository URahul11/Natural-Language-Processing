```python
pos_dict={
    "cat":"NOUN",
    "dog":"NOUN",
    "child":"NOUN",
    "run":"VERB",
    "eat":"VERB",

    "play":"VERB",
    "happy":"ADJECTIVE",
    "blue":"ADJECTIVE",
    "i":"PRONOUN",
    "you":"PRONOUN",
}
word=input().lower()
if word in pos_dict:
  print(pos_dict[word])
else:
  print("UNKNOWN")
```
